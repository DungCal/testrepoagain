```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('C:/Users/admin/Desktop/lake-192990__340.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
def color_balance(contrast,brightness):
    img_=img.copy()
    img_= (img_*contrast+brightness).astype('int')
    img_= np.clip(img_,0,255)
    return img_
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,7))

ax[0].imshow(img)
ax[0].set_title('Original')
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].imshow(color_balance(1.2,30))
ax[1].set_title('After balance')
ax[0].set_xticks([])
ax[0].set_yticks([])

plt.show()

```

Histogram Equalization

```python
def equal_hist(img):
    img__=img.copy()
    for i in range(img__.shape[2]):
        img__[:,:,i]=cv2.equalizeHist(img__[:,:,i])
    return img__
```

```python
fig,ax=plt.subplots(2,2,figsize=(15,7))
ax[0,0].imshow(color_balance(1.2,30))
ax[0,0].set_title('Original')
ax[0,0].set_xticks([])
ax[0,0].set_yticks([])

histr= cv2.calcHist([np.uint8(color_balance(1.2,30))],[0],None,[256],[0,255])
histr_1= cv2.calcHist([np.uint8(color_balance(1.2,30))],[1],None,[256],[0,255])
histr_2= cv2.calcHist([np.uint8(color_balance(1.2,30))],[2],None,[256],[0,255])
ax[1,0].plot(histr,c='b')
ax[1,0].plot(histr_1,c='g')
ax[1,0].plot(histr_2,c='r')

ax[0,1].imshow(equal_hist((np.uint8(color_balance(1.2,30)))))
ax[0,1].set_title('Equalized')
ax[0,1].set_xticks([])
ax[0,1].set_yticks([])

histr= cv2.calcHist([equal_hist(np.uint8(color_balance(1.2,30)))],[0],None,[256],[0,255])
histr_1= cv2.calcHist([equal_hist(np.uint8(color_balance(1.2,30)))],[1],None,[256],[0,255])
histr_2= cv2.calcHist([equal_hist(np.uint8(color_balance(1.2,30)))],[2],None,[256],[0,255])
ax[1,1].plot(histr,c='b')
ax[1,1].plot(histr_1,c='g')
ax[1,1].plot(histr_2,c='r')
```

### Function 3: implement the median filter to remove noise in the image(salt and pepper noise)

```python
def medianblur():
    from skimage.util import random_noise

    # Load an image
    im_arr = cv2.imread('C:/Users/admin/Desktop/lake-192990__340.jpg')

    # Add salt and pepper noise to the image
    noise_img = random_noise(im_arr, mode="s&p",amount=0.3)
    noise_img = np.array(255*noise_img, dtype = 'uint8')


    noise_img=cv2.cvtColor(noise_img,cv2.COLOR_BGR2RGB)
    # Apply median filter
    median = cv2.medianBlur(noise_img,5)

    # Display the image
    fig,ax=plt.subplots(1,2,figsize=(10,7))
    ax[0].imshow(noise_img)
    ax[0].set_title('Image with Salt and pepper noises')
    ax[0].set_xticks([])
    ax[0].set_yticks([])

    ax[1].imshow(median)
    ax[1].set_title('Median Blur applied')
    ax[1].set_xticks([])
    ax[1].set_yticks([])

    plt.show()
# cv2.imshow('blur',noise_img)
# cv2.imshow('blur1',median)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
medianblur()
```

### Function 4: implement the Mean filter to remove noise in image (salt and pepper noise)

```python
#import numpy as np

def meanblur():
    from skimage.util import random_noise

    # Load an image
    im_arr = cv2.imread('C:/Users/admin/Desktop/lake-192990__340.jpg')

    # Add salt and pepper noise to the image
    noise_img = random_noise(im_arr, mode="s&p",amount=0.3)
    noise_img = np.array(255*noise_img, dtype = 'uint8')


    noise_img=cv2.cvtColor(noise_img,cv2.COLOR_BGR2RGB)
    # Apply mean filter
    median = cv2.blur(noise_img,(5,5))

    # Display the image
    fig,ax=plt.subplots(1,2,figsize=(10,7))
    ax[0].imshow(noise_img)
    ax[0].set_title('Image with Salt and pepper noises')
    ax[0].set_xticks([])
    ax[0].set_yticks([])

    ax[1].imshow(median)
    ax[1].set_title('Mean Blur applied')
    ax[1].set_xticks([])
    ax[1].set_yticks([])

    plt.show()
    # cv2.imshow('blur',noise_img)
    # cv2.imshow('blur1',median)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()
```

### Function 5: implement Gaussian smoothing to perform image smoothing.

```python
#import numpy as np
def gaussian():
    from skimage.util import random_noise

    # Load an image
    im_arr = cv2.imread('C:/Users/admin/Desktop/lake-192990__340.jpg')

    # Add salt and pepper noise to the image
    noise_img = random_noise(im_arr, mode="gaussian")
    noise_img = np.array(255*noise_img, dtype = 'uint8')


    noise_img=cv2.cvtColor(noise_img,cv2.COLOR_BGR2RGB)
    # Apply mean filter
    median = cv2.GaussianBlur(noise_img,(3,3),0)

    # Display the image
    fig,ax=plt.subplots(1,2,figsize=(10,7))
    ax[0].imshow(noise_img)
    ax[0].set_title('Image with Gaussian noises')
    ax[0].set_xticks([])
    ax[0].set_yticks([])

    ax[1].imshow(median)
    ax[1].set_title('Gaussian Blur applied')
    ax[1].set_xticks([])
    ax[1].set_yticks([])

    plt.show()
    # cv2.imshow('blur',noise_img)
    # cv2.imshow('blur1',median)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()
```

```python
gaussian()
```
